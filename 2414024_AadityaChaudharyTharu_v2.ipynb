# Task - 3 - Outlier Detection:

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule

```python
Q1_score = ranked_sac_dataset['score'].quantile(0.25)
Q3_score = ranked_sac_dataset['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = ranked_sac_dataset['Log GDP per capita'].quantile(0.25)
Q3_gdp = ranked_sac_dataset['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

outliers_score = ranked_sac_dataset[
    (ranked_sac_dataset['score'] < lower_bound_score) |
    (ranked_sac_dataset['score'] > upper_bound_score)
]

outliers_gdp = ranked_sac_dataset[
    (ranked_sac_dataset['Log GDP per capita'] < lower_bound_gdp) |
    (ranked_sac_dataset['Log GDP per capita'] > upper_bound_gdp)
]

outliers = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

outliers[['Country name', 'score', 'Log GDP per capita']]
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(ranked_sac_dataset['Log GDP per capita'], ranked_sac_dataset['score'], label='Countries', c='blue', alpha=0.7)

plt.scatter(outliers['Log GDP per capita'], outliers['score'], label='Outliers', c='green', alpha=1)

plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('Scatter Plot: GDP per Capita vs. Score')

plt.legend()

plt.show()
```
