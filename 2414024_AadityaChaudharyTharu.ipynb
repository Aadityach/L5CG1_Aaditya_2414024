# Task - 3 - Outlier Detection:

Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python

# Calculate IQR for Composite Score and GDP per capita
Q1_composite = sac_dataset['Composite Score'].quantile(0.25)
Q3_composite = sac_dataset['Composite Score'].quantile(0.75)
IQR_composite = Q3_composite - Q1_composite
lower_bound_composite = Q1_composite - 1.5 * IQR_composite
upper_bound_composite = Q3_composite + 1.5 * IQR_composite

Q1_gdp = sac_dataset['Log GDP per capita'].quantile(0.25)
Q3_gdp = sac_dataset['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers for Composite Score and GDP per capita
outliers_composite = sac_dataset[(sac_dataset['Composite Score'] < lower_bound_composite) |
                                 (sac_dataset['Composite Score'] > upper_bound_composite)]

outliers_gdp = sac_dataset[(sac_dataset['Log GDP per capita'] < lower_bound_gdp) |
                           (sac_dataset['Log GDP per capita'] > upper_bound_gdp)]

# Combine outliers from both criteria
outliers = pd.concat([outliers_composite, outliers_gdp]).drop_duplicates()

# Display outliers
print("Outliers based on Composite Score and GDP per Capita:")
print(outliers[['Country name', 'Composite Score', 'Log GDP per capita']])
```

```python
# Calculate IQR for Composite Score and GDP per capita using the 1.5 * IQR rule
Q1_composite = sac_dataset['Composite Score'].quantile(0.25)
Q3_composite = sac_dataset['Composite Score'].quantile(0.75)
IQR_composite = Q3_composite - Q1_composite
lower_bound_composite = Q1_composite - 1.5 * IQR_composite
upper_bound_composite = Q3_composite + 1.5 * IQR_composite

Q1_gdp = sac_dataset['Log GDP per capita'].quantile(0.25)
Q3_gdp = sac_dataset['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers for Composite Score and GDP per capita based on the IQR rule
outliers_composite = sac_dataset[(sac_dataset['Composite Score'] < lower_bound_composite) |
                                 (sac_dataset['Composite Score'] > upper_bound_composite)]

outliers_gdp = sac_dataset[(sac_dataset['Log GDP per capita'] < lower_bound_gdp) |
                           (sac_dataset['Log GDP per capita'] > upper_bound_gdp)]

# Combine the outliers from both criteria
outliers = pd.concat([outliers_composite, outliers_gdp]).drop_duplicates()

# Display the outliers
print("Outliers based on 1.5 × IQR rule for Composite Score and GDP per Capita:")
print(outliers[['Country name', 'Composite Score', 'Log GDP per capita']])
```
